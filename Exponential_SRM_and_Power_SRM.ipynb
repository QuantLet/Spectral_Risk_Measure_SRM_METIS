```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import cauchy
from scipy.stats import uniform
from scipy.stats import beta
from scipy.stats import gumbel_r
from scipy.stats import t
```

# Exponential SRM

```python
# Exponential SRM with normal distribution
plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':75})
n = 500
draws = norm.rvs(size = n)
p = np.linspace(0,1,n)
K = range(0,100)
M_phi_norm = np.zeros(100)

for k in range(1,100):
   sum_uz = 0
   sum_uz_final = 0
   for i in range(len(p)):
        #z = z_vals[i]  #equal to quantile 'ndtri(p)'
        #q = scipy.stats.norm.cdf(p[i], loc=0.0, scale=1.0)
        q = np.quantile(draws, p[i])
        u = np.exp(-k*(1-p[i])) 
        z = q
        uz= u*z
        sum_uz= sum_uz+uz
        #print(i, "z = %0.8f" % z, "p = %0.8f" % p, "u = %0.8f" % u,"uz = %0.8f" % uz,"sum_uz = %0.8f" % sum_uz,"sum_uz_final = %0.8f" % sum_uz_final)
   sum_uz_final= k/(1-np.exp(-k))/n*sum_uz
   M_phi_norm[k] = sum_uz_final
   #print(sum_uz_final)
plt.plot(K,M_phi_norm)
```

#Power SRM with gamma <1

```python
plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':75})
n = 500
draws = norm.rvs(size = n)
p = np.linspace(0,0.99,n)
Gamma = np.arange(0.01,1,0.01)
M_phi_power_norm = np.zeros(99)
j=0
for gamma in Gamma:
   sum_uz = 0
   sum_uz_final = 0
   
   for i in range(0,len(p)):
        q = np.quantile(draws, p[i])
        u = gamma*(1-p[i])**(gamma-1) 
        z = q
        uz= u*z
        sum_uz= sum_uz+uz
        sum_uz_final= 1/float(n)*sum_uz
   
   M_phi_power_norm[j] = sum_uz_final
   j+=1
   #print(sum_uz_final)
plt.plot(Gamma,M_phi_power_norm)
```

#Power SRM with gamma > 1

```python
plt.rcParams.update({'figure.figsize':(8,5), 'figure.dpi':75})
n = 500
draws = norm.rvs(size = n)
p = np.linspace(0,1,n)
Gamma = range(0,100)
M_phi_power_norm = np.zeros(100)

for gamma in range(1,100):
   sum_uz = 0
   sum_uz_final = 0
   for i in range(len(p)):
        q = np.quantile(draws, p[i])
        u = gamma*p[i]**(gamma-1) 
        z = q
        uz= u*z
        sum_uz= sum_uz+uz
        sum_uz_final= 1/n*sum_uz
   M_phi_power_norm[gamma] = sum_uz_final
   #print(sum_uz_final)
plt.plot(Gamma,M_phi_power_norm)
```
